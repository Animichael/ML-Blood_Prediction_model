Importing Libraries for the model

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import pickle
```

Data Loading and Preprocessing

```python
blood_Pressure_Dataset =pd.read_csv('/content/Blood Pressure Dataset.csv')
```

```python
# Checking the first 10 rows of the dataset
blood_Pressure_Dataset.head(20)

```

Getting more information about the dataset

```python
blood_Pressure_Dataset.info()
```

```python
# Checking for missing values in the dataset
blood_Pressure_Dataset.isnull().sum().sum()
```

```python
blood_Pressure_Dataset.tail(10)
```

```python
fill_data = blood_Pressure_Dataset.fillna(value=1)
fill_data.isnull().sum()
```

```python
# Checking the Statiscal information about the dataset
fill_data.describe()

```

```python
#  checking the size of the dataset after deleting the missing values
fill_data.shape
```

```python
# checking for the values of each outcome
fill_data['result'].value_counts()
```

```python
# Lable 0 --> People with no high blood pressure
# Lable 1 --> People with  high blood

# Finding the mean of these two outcomes

fill_data.groupby('result').mean()
```

Separating the independent variable from the dependent variables

```python
x= fill_data.drop(columns='result', axis=1)
y= fill_data['result']

# x= fill_data.drop('result', axis=1)
# y=fill_data
```

```python
# printing the values stored in x variable
print(x)
```

```python
# printing the values of the target variable y
print(y)
```

Data standardization (Data Encoding)

```python
scaler = StandardScaler()

scaler.fit(x)
```

```python
# Transforming the data
standardize_Data = scaler.transform(x)
```

```python
# Printing the standardized data
print(standardize_Data)
```

```python
x= standardize_Data
y= fill_data['result']
```

```python
print(x)
print(y)
```

Splitting our data into train and test data

```python
x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.1, stratify=y, random_state=2)
```

```python
# Checking the shape of our origin x, x_train and x_test
print(x.shape, x_train.shape, x_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

Training the model

 Using Support Vector Classifier


```python
classifier =svm.SVC(kernel='linear')
```

```python
# training the support vector Machine classifier
classifier.fit(x_train, y_train)
```

```python
# Model Evaluation
```

 1. Accuracy Score For SVM model

```python
# The Accuracy on the training dataset
x_train_prediction =classifier.predict(x_train)
training_Data_Accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
# printing the accuracy score of the training data
print('The Accuracy Score of the training data is:',training_Data_Accuracy )
```

```python
# The Accuracy on the test data
x_test_prediction =classifier.predict(x_test)
test_Data_Accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
# printing the accuracy score of the test data
print('The Accuracy Score of the test data is:',test_Data_Accuracy )
```

Developing the predictive system

```python
input_data =(29,1,9.37,0.95,38,11,1,0,0,30738,48217,336,1,1,1)
# changing the input data to a numpy array
input_data_as_array =np.asarray(input_data)

# reshaping the arraay as we are predicting for one instance
input_data_reshape = (input_data_as_array.reshape(1,-1))

# Standardizing the input data

std_data = scaler.transform(input_data_reshape)
print(std_data )

# Making the prediction
prediction = classifier.predict(std_data)
print(prediction)

if(prediction[0]==0):
   print('The person is not having High blood pressure')
else:
  print('The person is having High blood pressure')
print()
print()
```
